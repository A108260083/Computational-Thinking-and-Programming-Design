```
while True:
     acc = input('請輸入帳號？')
     pwd = input('請輸入密碼？')
     if (acc == 'abc' and pwd == '123'):
        print('帳號與密碼正確')
        break
     else:
         print('登入失敗')
```

```
password = "a123456"
times = 0

while times <= 3:
    answer = input("Enter Your Password: ") #另外⽤⼀個變數名字
    if answer == password: #不要⽤a123456，因為密碼可以改
       print("登入成功")
       break #出迴圈
    else:
         times = times + 1
         chance = 3 - times
         print("密碼錯誤！還有", chance, "次機會")
```

```
s = int(input('請輸入加總開始值？'))
e = int(input('請輸入加總終⽌值？'))
inc= int(input('請輸入遞增減值？'))

sum = 0   #初始結果

for i in range(s, e, inc): #判斷條件

    sum = sum + i #更新結果
    print('i為', i, '加總結果為', sum)
```
